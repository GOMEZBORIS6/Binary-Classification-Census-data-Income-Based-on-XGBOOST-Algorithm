## Import Libraries

```python
# import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
# import os
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score
import xgboost as xgb
import itertools
%matplotlib inline 
import matplotlib.pyplot as plt
```

## Loading Data

As the dataset is hosted directly on the [**UCI Machine Learning website**](https://archive.ics.uci.edu/dataset/20/census+income), it is conveniently accessible for direct download from the internet.

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', names=['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                              'marital_status', 'occupation', 'relationship', 'race', 'sex',
                              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country',
                              'income'], sep=', ', engine="python", encoding='latin-1')
df.head()
```

```python
census_df = df.copy()
census_df.head()
```

We note that our dataset includes individuals from **42** different countries. Training our model on all these data may introduce biases into our results. One potential bias is **geographical bias**, meaning that the data may be influenced by countries, as individuals characteristics and behaviors may vary according to culture, history, the economic and political environment, as well as other country-related factors. These elements are beyond our control with this data set. To resolve this, we will group our dataset into two categories: **US-native individuals**, as they represent the bulk of our data, and **non-US-native individuals**. This approach also dispenses with the need to perform a **one-hot encoding** for a variable with **42** modalities, allowing us instead to recode the variable by assigning the value 1 for the USA and 0 for other countries.

```python
# n_country = 1 if United-States
# n_country = 0 otherwise
n_country = []
for element in census_df["native_country"]:
    if element == "United-States":
        n_country.append(1)
    else:
        n_country.append(0)
        

census_df['n_country']=n_country
census_df.drop('native_country', inplace=True, axis=1)
census_df.head()
```

we encode our **target** variable, i.e. the dependent variable **income**. It takes the value 1 if the income is greater than 50K and 0 otherwise. 

```python
income_dum = pd.get_dummies(census_df['income'], prefix='income', drop_first=True)
census_df['income_50K']=income_dum
census_df.drop('income', inplace=True, axis=1)
census_df.head()
```

## Split target and covariates variables

```python
# Split into dependend and independent variables
X = census_df.drop('income_50K', axis=1)
y = census_df['income_50K']
```

```python
# Split X into continous variables and categorical variables
# Notice Native_country variable is already encoding as n_country

X_continous  = X[['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']]

X_categorical = X[['workclass', 'marital_status', 'occupation', 'relationship', 'race','sex', "n_country"]]
```

```python
# Get the dummies
X_encoded = pd.get_dummies(X_categorical, drop_first=True)
# Concatenate both continous and encoded sets:
X = pd.concat([X_continous, X_encoded],axis=1)
X.head()
```

```python
X.loc[X['capital_loss']>0, 'capital_loss'] = 1
X.loc[X['capital_gain']>0, 'capital_gain'] = 1
X.head()
```

```python
X.isnull().sum()
```

## Normalisation 

```python
transform = preprocessing.StandardScaler()
X = transform.fit(X).transform(X.astype(float))
X
```

## Train/Test Split

```python
## Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 1/3,
                                                    stratify=y, random_state=10)
```

```python
print("Shape of X_train data : ",X_train.shape)
print("Shape of X_test data : ",X_test.shape)
print("Shape of y_train data : ",y_train.shape)
print("Shape of y_test data : ",y_test.shape)
```

## **XGBoost Application**

Now, our focus shifts to assessing the performance of XGBoost. Initially, we establish a variable, `learning_rate_range`, to store the range of learning rates we intend to experiment with, spanning from **0.01** to **1** with an interval of **0.05**. Subsequently, the training and testing scores of XGBoost are recorded in the lists `train_Xgb` and `test_Xgb`.

```python
learning_rate_range = np.arange(0.01, 1, 0.05)
test_Xgb = [] 
train_Xgb = []
for lr in learning_rate_range:
    xgb_classifier = xgb.XGBClassifier(eta = lr)
    xgb_classifier.fit(X_train, y_train)
    train_Xgb.append(xgb_classifier.score(X_train, y_train))
    test_Xgb.append(xgb_classifier.score(X_test, y_test))
```

we can draw a line plot to see how XGBoost performs.

```python
fig = plt.figure(figsize=(10, 7))
plt.plot(learning_rate_range, train_Xgb, c='blue', label='Train')
plt.plot(learning_rate_range, test_Xgb, c='red', label='Test')
plt.xlabel('Learning rate')
plt.xticks(learning_rate_range)
plt.ylabel('Accuracy score')
plt.ylim(0.6, 1)
plt.legend(prop={'size': 12}, loc=3)
plt.title('Accuracy score vs. Learning rate of XGBoost', size=14)
plt.show()
```

XGBoost did an okay job. It has **84%** accuracy when learning rate is **0.06**.
Even if the model exhibits satisfactory performance on the current testing data, there is no assurance that it will consistently maintain such performance.
To address this concern, regularization becomes imperative. By applying regularization to the model, we intentionally limit its capacity, mitigating the risk of overfitting issues.

In general, a reduction in the number of iterations (`n_estimators`) and learning rate (`learning_rate`), or an increase in the minimum gain required in a node (`gamma`) and the regularization parameter (`reg_lambda`) can be employed to prevent the model from excessively capturing the nuances of the training set.

For this particular instance, we performing a **Grid search** to determine the optimal  `reg_lambda`, `max_depth`, `gamma`, `subsample` and `min_child_weight`, aiming to mitigate the risk of overfitting. In real-world scenarios, **Grid search** coupled with **K-fold cross-validation** is often employed to identify the most effective combination of parameters.

Additionally, it's noteworthy that as the `learning rate` surpasses **0.5**, the training score tends to approach **0.95**, while the testing score steadily **declines**, signaling an evident issue of overfitting. To address this, we adjust **the learning_rate_range** to span from **0.01** to **0.5** with an interval of **0.05**.

```python
param_grid = {
    'max_depth': [3, 4, 5, 7],
    'learning_rate': np.arange(0.01, 0.5, 0.05),
    'subsample': [0.4, 0.8, 1],
    'gamma': [0, 1, 10],
    'reg_lambda': [0, 1, 10]
}

cv = StratifiedKFold(n_splits=3)

optimal_params = GridSearchCV(
    estimator=xgb.XGBClassifier(objective='binary:logistic',
                                eval_metric="logloss",
                                use_label_encoder=False), #silent =1 for don't show a message
    param_grid=param_grid,
    scoring='accuracy',
    verbose=1,
    cv=cv
)
optimal_params.fit(X_train, y_train)

print("Best hyperparameters : ",optimal_params.best_params_)
print("Accuracy :",optimal_params.best_score_)



#Best hyperparameters :  {'gamma': 1, 'learning_rate': 0.16, 'max_depth': 4, 'reg_lambda': 1, 'subsample': 0.8}
#Accuracy : 0.8508314031470535
```

```python
learning_rate_range = np.arange(0.01, 0.5, 0.05)
fig = plt.figure(figsize=(19, 17))
idx = 1
# grid search for min_child_weight
for weight in np.arange(0, 4.5, 0.5):
    train = []
    test = []
    for lr in learning_rate_range:
        xgb_classifier = xgb.XGBClassifier(eta = lr, objective='binary:logistic', eval_metric="logloss", use_label_encoder=False,
                                           reg_lambda=1, min_child_weight=weight,
                                           subsample=0.8, gamma = 1, max_depth = 4)
        xgb_classifier.fit(X_train, y_train)
        train.append(xgb_classifier.score(X_train, y_train))
        test.append(xgb_classifier.score(X_test, y_test))
    fig.add_subplot(3, 3, idx)
    idx += 1
    plt.plot(learning_rate_range, train, c='blue', label='Training')
    plt.plot(learning_rate_range, test, c='red', label='Testing')
    plt.xlabel('Learning rate')
    plt.xticks(learning_rate_range)
    plt.ylabel('Accuracy score')
    plt.ylim(0.6, 1)
    plt.legend(prop={'size': 12}, loc=3)
    title = "Min child weight:" + str(weight)
    plt.title(title, size=16)
plt.show()
```

As you can see, the over-fitting problem is somewhat serious when `min_child_weight` is equal to **1.5** and **2.5**, whereas training and test scores generally become closer as `min_child_weight` **increases**. More specifically, the over-fitting problem seems minor when `min_child_weight` is greater than or equal to **3.0**.

```python
# MODEL with minor overffiting problem and bests parameters
xgb_model = xgb.XGBClassifier(eta = 0.16, objective='binary:logistic', eval_metric="logloss", use_label_encoder=False,
                                           reg_lambda=1, min_child_weight=3.0,
                                           subsample=0.8, gamma = 1, max_depth = 4)

xgb_model = xgb_model.fit(X_train, y_train)

# CROSS VALIDATION
cv = StratifiedKFold(n_splits=3)
val_xgb_model = cross_val_score(xgb_model, X_train, y_train, cv=cv).mean()
val_xgb_model
```

```python
learning_rate_range = np.arange(0.01, 0.5, 0.05)
fig = plt.figure(figsize=(10, 6))

train = []
test = []
for lr in learning_rate_range:
    xgb_classifier = xgb.XGBClassifier(eta = lr, objective='binary:logistic', eval_metric="logloss", use_label_encoder=False,
                                        reg_lambda=1, min_child_weight=3.0,
                                        subsample=0.8, gamma = 1, max_depth = 4)
    xgb_classifier.fit(X_train, y_train)
    train.append(xgb_classifier.score(X_train, y_train))
    test.append(xgb_classifier.score(X_test, y_test))

plt.plot(learning_rate_range, train, c='blue', label='Training')
plt.plot(learning_rate_range, test, c='red', label='Testing')
plt.xlabel('Learning rate')
plt.xticks(learning_rate_range)
plt.ylabel('Accuracy score')
plt.ylim(0.6, 1)
plt.legend(prop={'size': 12}, loc=3)
title = "Min child weight: 3.0"
plt.title(title, size=16)
plt.show()
```

```python
#This function is to plot the confusion matrix.
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    np.set_printoptions(precision=5)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['income_<_50K', 'income_>_50K']); ax.yaxis.set_ticklabels(['income_<_50K', 'income_>_50K']) 
    plt.show() 
```

```python
# PREDICTIONS
xgbm_predictions = xgb_model.predict(X_test)
plot_confusion_matrix(y_test,xgbm_predictions)
```
